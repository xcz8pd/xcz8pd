```python
import pandas as pd

acc_stats_df = pd.read_csv('acc_players-2324F.csv', header=1)
print("Total 2023-2024 ACC Stats")
acc_stats_df.head(5)
```

```python
total_points = acc_stats_df['PTS'].sum()
print("Total points scored:")
total_points
```

```python
most_minutes = acc_stats_df[['Player','MP']].sort_values(by='MP', ascending=False)
print("Player with most minutes:")
most_minutes.head(1)
```

```python
top_5_trb = acc_stats_df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)
print("Top 5 players with most rebounds:")
top_5_trb
```

```python
over_500_minutes = acc_stats_df[acc_stats_df['MP'] > 500]
print("Players with more than 500 minutes:")
over_500_minutes
```

```python
highest_ast = over_500_minutes[['Player','AST']].sort_values(by='AST', ascending=False)
print("Player with most assists:")
highest_ast.head(1)
```

```python
top_3_assist = acc_stats_df[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)
print("Top 3 players with most assists:")
top_3_assist
```

```python
top_3_shot_blockers = acc_stats_df[['Player', 'BLK']].sort_values(by='BLK', ascending=False).head(3)
print("Top 3 players with most blocks:")
top_3_shot_blockers
```

```python
total_points_per_school = acc_stats_df.groupby('School')[['PTS']].sum()
print("Total points scored per school:")
total_points_per_school
```

```python
total_assists_per_school = acc_stats_df.groupby('School')[['AST']].sum()
print("Total assists per school:")
total_assists_per_school
```

```python
total_points_per_school.sort_values(by='PTS', ascending=False).head(3)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

top_5_scorers = acc_stats_df[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(5)

plt.figure(figsize=(10,6))
sns.barplot(x='PTS', y='Player', data=top_5_scorers)
plt.xlabel('Points')
plt.ylabel('Player')
plt.title('Top 5 2023-2024 Season Scorers')
plt.show()
```

```python
plt.scatter(acc_stats_df['FG%'], acc_stats_df['WS'], color='blue')
plt.xlabel('Field Goal %')
plt.ylabel('Win Shares')
plt.title('Scatter Plot of Field Goal % vs. Win Shares')
plt.show()

correlation = acc_stats_df['FG%'].corr(acc_stats_df['WS'])
print(f"Correlation coefficient: {correlation}")
```
