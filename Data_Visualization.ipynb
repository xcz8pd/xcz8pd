```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

covid = pd.read_csv('covid.csv')

county_deaths = covid.groupby('State')['COVID-19 Deaths'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))  # Adjust figure size as needed
plt.bar(county_deaths.index, county_deaths.values)
plt.xlabel("State")
plt.ylabel("Total COVID Deaths")
plt.title("COVID Deaths by State")
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

aqi = pd.read_csv('aqi_2020.csv')

aqi_Unhealthy = aqi.groupby('State')['Unhealthy Days'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))  # Adjust figure size as needed
plt.bar(aqi_Unhealthy.index, aqi_Unhealthy.values)
plt.xlabel("State")
plt.ylabel("Total Unhealthy Days")
plt.title("Total Unhealthy Days by State")
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Load the Gapminder-like dataset
aqi = pd.read_csv('aqi_2020.csv')
aqi_Unhealthy = aqi.groupby('State')['Unhealthy Days'].sum().sort_values(ascending=False)
aqi_Unhealthy = aqi_Unhealthy.reset_index()

# Create an interactive scatter plot
fig = px.scatter(
    aqi_Unhealthy,
    x='State',
    y='Unhealthy Days',
    size='Unhealthy Days',  # Bubble size represents population
    color='State',
    title='Unhealthy Days by State'
)
fig.show()
```
