```python
from pymongo import MongoClient
import pprint

connection_string = "mongodb+srv://claregibb:31Grantpl@cluster0.by451.mongodb.net/test?retryWrites=true&w=majority"

client = MongoClient(connection_string)

db = client['sample_mflix']
collection = db['movies']
```

```python
action_movies = collection.find({"genres": "Action"}).sort("year", 1).limit(1)
for movie in action_movies:
    pprint.pprint(movie)
```

```python
movies_after_2000 = collection.find({"year": {"$gt": 2000}}).limit(5)
for movie in movies_after_2000:
    pprint.pprint(movie)
```

```python
high_rating_movies = collection.find({"imdb.rating": {"$gt": 8.5}}).limit(5)
for movie in high_rating_movies:
    pprint.pprint(movie)
```

```python
action_adventure_movies = collection.find({"genres": {"$all": ["Action", "Adventure"]}}).limit(5)
for movie in action_adventure_movies:
    pprint.pprint(movie)
```

```python
comedy_movies = collection.find({"genres": "Comedy"}).sort("imdb.rating", -1).limit(5)
for movie in comedy_movies:
    pprint.pprint(movie)
```

```python
drama_movies = collection.find({"genres": "Drama"}).sort("year", 1).limit(5)
for movie in drama_movies:
    pprint.pprint(movie)
```

```python
avg_genre_rating = [
    {"$unwind": "$genres"},
    {"$group": {"_id": "$genres", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"avg_rating": -1}},
    {"$limit": 5}
]

aggregated_data = collection.aggregate(avg_genre_rating)

for data in aggregated_data:
    pprint.pprint(data)
```

```python
top_directors = collection.aggregate([
    {"$group": {"_id": "$directors", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"avg_rating": -1}},
    {"$limit": 5}
])

for data in top_directors:
    pprint.pprint(data)
```

```python
movies_per_year = collection.aggregate([
    {"$group": {"_id": "$year", "count": {"$sum": 1}}},
    {"$sort": {"_id": 1}}
])

for data in movies_per_year:
    pprint.pprint(data)
```

```python
collection.update_one({"title": "The Godfather"}, {"$set": {"imdb.rating": 9.5}})
```

```python
collection.update_many({"genres": "Horror", "imdb.rating": {"$exists": False}}, {"$set": {"imdb.rating": 6.0}})
```

```python
collection.delete_many({"year": {"$lt": 1950}})
```

```python
##collection.create_index([("title", "text")])
movies_with_love = collection.find({"$text": {"$search": "love"}})

##collection.create_index([("title", "text"), ("plot", "text")])
war_movies = collection.find({"$text": {"$search": "war"}}).sort("imdb.rating", -
1).limit(5)
for movie in war_movies:
    pprint.pprint(movie)
```

```python
multiple_conditions = {"genres": "Action", "imdb.rating": {"$gt": 8}}
results = collection.find(multiple_conditions).sort("year", -1).limit(5)
for result in results:
    pprint.pprint(result)
```

```python
multi_condition = {"directors": "Christopher Nolan", "imdb.rating": {"$gt": 8}}
nolan_movies = collection.find(multi_condition).sort("imdb.rating", -1).limit(3)
for movie in nolan_movies:
    pprint.pprint(movie)
```
