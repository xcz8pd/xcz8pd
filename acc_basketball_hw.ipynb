```python
import pandas as pd

acc_stats_df = pd.read_csv('acc_players-2324F.csv', header=1)
print("Total 2023-2024 ACC Stats")
acc_stats_df.head(5)
```

```python
total_points = acc_stats_df['PTS'].sum()
print("Total points scored:")
total_points
```

```python
most_minutes = acc_stats_df[['Player','MP']].sort_values(by='MP', ascending=False)
print("Player with most minutes:")
most_minutes.head(1)
```

```python
top_5_trb = acc_stats_df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)
print("Top 5 players with most rebounds:")
top_5_trb
```

```python
over_500_minutes = acc_stats_df[acc_stats_df['MP'] > 500]
print("Players with more than 500 minutes:")
over_500_minutes
```

```python
highest_ast = over_500_minutes[['Player','AST']].sort_values(by='AST', ascending=False)
print("Player with most assists:")
highest_ast.head(1)
```
